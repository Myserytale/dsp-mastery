## Exercise
For the transfer function $H(z) = \frac{1}{z-0.5}$, determine whether the system is stable. Justify your answer.

### Solution 1


$$
H(z)=\frac{z^{-1}}{1-0.5z^{-1}}.
$$

The pole is at

$$
z=0.5.
$$

The algebraic expression alone does not determine the impulse response, because the ROC is not specified.

If the system is causal, then

$$
\mathrm{ROC}: |z|>0.5.
$$

Then

$$
H(z)=z^{-1}\frac{1}{1-0.5z^{-1}}
=
z^{-1}\sum_{n=0}^{\infty}(0.5)^n z^{-n}
$$

so

$$
H(z)=\sum_{n=0}^{\infty}(0.5)^n z^{-(n+1)}.
$$

Therefore,

$
h[n]=(0.5)^{n-1}u[n-1] = \{0, 1, 0.5, 0.25, \dots, 0.5^{n},\dots\}\ .
$
<img src="Meddins_zplane_3.png" width=500>

Now test absolute summability:

$$
\sum_{n=-\infty}^{\infty}|h[n]| = \sum_{n=1}^{\infty}(0.5)^{n-1} = 1+0.5+0.25+\cdots = 2<\infty.
$$

Thus, if the system is causal, it is stable.

However, if the ROC is

$$
|z|<0.5,
$$

then

$$
H(z)=\frac{1}{z-0.5}
=
-\frac{2}{1-2z}
=
-2\sum_{n=0}^{\infty}(2z)^n
$$

so

$$
H(z)=
-\sum_{n=0}^{\infty}2^{n+1}z^n.
$$

Since

$$
z^n=z^{-(-n)},
$$

this corresponds to

$$
h[-n]=-2^{n+1},
\qquad n\ge 0.
$$

Equivalently,

$$
h[n]=-2^{1-n}u[-n].
$$

This is not absolutely summable:

$$
\sum_{n=-\infty}^{0}|h[n]|=\infty.
$$

Therefore,

$$
\text{Stable for ROC } |z|>0.5,\quad
\text{unstable for ROC } |z|<0.5.
$$

### Remark

There is another way to get the impulse response. 
We know from theory that the most general form of a causal LTI is:
	$$
	\sum_{l=0}^m b_l\cdot y[n-l] = \sum_{l=0}^k a_l\cdot x[n-l]
	$$
	which at the level of the $z$-transforms
	$$
	B(z)\cdot Y(z) = A(z)\cdot X(z)\quad\rightarrow\quad
	Y(z) = \dfrac{A(z)}{B(z)} X(z)\ \quad\rightarrow\quad H(z) = \dfrac{A(z)}{B(z)}\ .
	$$

In our case $A(z) = z^{-1}$, $B(z) = 1 - 0.5z^{-1}$ whence $a[n] = \delta[n-1]=\{0, 1\}$ and $b[n]=\delta[n] - 0.5\delta[n-1]=\{1, -0.5\}$.

Therefore the system is defined as $y[n]-0.5y[n-1] = x[n-1]$ or $y[n]=0.5y[n-1] + x[n-1]$.

For the impulse response: $x[n] = \delta[n]$, $y[n]=h[n]$.

Thus, $h[n] = 0.5h[n-1] + \delta[n-1]$.

That is:

$h[0] = 0.5\cdot 0 + 0 = 0$

$h[1] = 0.5\cdot 0 + 1 = 1$

$h[2] = 0.5\cdot 1 + 0 = 0.5$

$h[3] = 0.5\cdot 0.5 + 0 = 0.25$

$h[n] = 0.5^{n-1}$

### Solution 2
Consider the input

$$
x[n]=e^{i\omega n}.
$$

(the sampling period $T$ is not specified therefore we work in time units of $T$ (equivalently assume that $T=1$) therefore $\omega T\rightarrow \omega\in[-\pi, \pi)$).

For an LTI system,

$$
y[n]=H(e^{i\omega})\,e^{i\omega n}.
$$

Evaluate the transfer function on the unit circle:

$$
H(e^{i\omega})=\frac{1}{e^{i\omega}-0.5}.
$$

Its magnitude is

$$
|H(e^{i\omega})|
=
\frac{1}{|e^{i\omega}-0.5|}.
$$

Since $e^{i\omega}$ lies on the unit circle, the minimum distance to $0.5$ is

$$
|1-0.5|=0.5,
$$

so

$$
|H(e^{i\omega})|\leq 2.
$$

Thus every sinusoidal input is scaled by a finite factor:

$$
|y[n]|\leq 2|x[n]|.
$$

This argument is valid only if the unit circle belongs to the ROC.

For the causal ROC

$$
|z|>0.5,
$$

the unit circle is included, so the system is stable.

## Exercise
Study the stability of the system with transfer function

$H(z) = \dfrac{1}{z^2 - z + 0.5}$		

and determine its impulse response.

### Solution
The two poles are $p_{1,2} = \frac{1\pm i}{2}$ $|p_{1,2}| < 1$ $\rightarrow$ stable causal system (the unit circle is within the ROC which is $|z| > \text{max}(|p_1|, |p_2|)$)<img src="Meddins_zplane_4.png" width=500>


$$
\frac{1}{z^2-z+0.5}
$$
$$
\begin{array}{rrrrrrrrrc|l}
1 & +0z & +0z^2 & +0z^3 & +0z^4 & +0z^5 & +0z^6 & +0z^7 & \cdots
&  &
z^2-z+0.5
\\\hline
&&&&&&&&
& &
z^{-2}+z^{-3}+0.5z^{-4}+0z^{-5}
-0.25z^{-6}-0.25z^{-7}
-0.125z^{-8}+0z^{-9}+\cdots
\\[1mm]
1 & -z & +0.5
\\
\hline
& z & -0.5
\\
& z & -z^2 & +0.5z
\\
\hline
&& 0.5z^2 & -0.5z
\\
&& 0.5z^2 & -0.5z^3 & +0.25z^2
\\
\hline
&&& -0.25z^2
\\
&&& -0.25z^2 & +0.25z & -0.125
\\
\hline
&&&& -0.25z & +0.125
\\
&&&& -0.25z & +0.25z^{-1} & -0.125z^{-2}
\\
\hline
&&&&& -0.125 & +0.125z^{-1}
\\
&&&&& -0.125 & +0.125z^{-1} & -0.0625z^{-2}
\\
\hline
&&&&&& 0.0625z^{-2}
\end{array}
$$


So $H(z)=z^{-2}+z^{-3}+0.5z^{-4}+0z^{-5}
-0.25z^{-6}-0.25z^{-7}
-0.125z^{-8}+0z^{-9}+\cdots$

That is $h[n] = \{0, 0, 1, 1, 0.5, 0, -0.25, -0.25, -0.125, 0, \dots\}$ <img src="Meddins_zplane_5.png" width=500>

### Alternative solution

$$
H(z) = \frac{z^{-2}}{1 - z^{-1} + 0.5z^{-2}}
$$
$A(z) = z^{-2}$, $B(z) = 1 - z^{-1} + 0.5z^{-2}$ $\rightarrow$ $a[n] = \{0, 0, 1\}, b[n] = \{1, -1, 0.5\}$

$y[n] - y[n-1] + 0.5y[n-2] = x[n-2]$ $\rightarrow$ $y[n] = y[n-1] - 0.5y[n-2] + x[n-2]$

$\rightarrow$ $h[n] = h[n-1] - 0.5h[n-2] + \delta[n-2]$

$h[0] = 0 - 0.5\cdot 0 + 0 = 0$

$h[1] = 0 - 0.5\cdot 0 + 0 = 0$

$h[2] = 0 - 0.5\cdot 0 + 1 = 1$

$h[3] = 1 - 0.5\cdot 0 + 0 = 1$

$h[4] = 1 - 0.5\cdot 1 + 0 = 0.5$

$h[5] = 0.5 - 0.5\cdot 1 + 0 = 0$

$h[6] = 0 - 0.5\cdot 0.5 + 0 = -0.25$

$h[7] = -0.25 - 0.5\cdot 0 + 0 = -0.25$

$h[8] = -0.25 - 0.5\cdot (-0.25) + 0 = -0.125$

$h[9] = -0.125 - 0.5\cdot (-0.25) + 0 = 0$

So

$h[n] = \{0, 0, 1, 1, 0.5, 0, -0.25, -0.25, -0.125, 0, \dots\}$

### Yet another solution
$p = p_1 = p_2^* = 1/\sqrt 2 e^{i\pi/4}$
$$
H(z) = \frac{1}{i}\left(\frac{1}{z - p}-\frac{1}{z - p^*}\right)=H_1(z) - H_2(z)
$$
From a previous exercise
$h_1[n] = (1/i)p^{n-1}u[n-1],\quad h_2[n] = (1/i)p^{*(n-1)}u[n-1]$

$h[n] = h_1[n] - h_2[n] = 2\text{Im}(p^{n-1}u[n-1])=2\left[\frac{\sin((n-1)\pi/4)}{(\sqrt 2)^{n-1}}\right]u[n-1]=\{0, 0, 1, 1, 0.5, 0, -0.25, -0.25, -0.125, 0, \dots\}$
