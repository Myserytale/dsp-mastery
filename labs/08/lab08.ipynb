## Gain and phase of continuous time system's transfer function: the Bode plot

Find the gain and phase angle for angular frequencies 0 rad/s (d.c.), 2 rad/s and for very large frequencies if the system is characterised by the transfer function
$$
				H(s) = \dfrac{s+1}{s+2}\rightarrow |H(j\omega)|=\sqrt{\dfrac{\omega^2 + 1}{\omega^2 + 4}}\ ,\quad \angle H(j\omega)=\arctan\frac{\omega}{1} - \arctan\frac{\omega}{2}
			$$
<img src="Meddins_splane_freqresponse_2.png" width="400">				
				
$\omega = 0$ rad/s $\rightarrow$ {Gain} = 0.5, phase angle = 0$^\circ$
				
$\omega = 2$ rad/s $\rightarrow$ {Gain} =$\sqrt{\frac{5}{8}}\approx 0,79$, phase angle = $\arctan 2 - \arctan 1 = 18.4^\circ$
				
$\omega \rightarrow\infty$ rad/s $\rightarrow$ {Gain} $\rightarrow$ 1, phase angle = $\frac{\pi}{2} - \frac{\pi}{2}=0^\circ$

```python
from pylab import *
from scipy import signal
num, den = [1, 1], [1, 2]
o, gain, phase = signal.bode((num, den))
def gain_theory(o):
    return 20*log10(sqrt(o**2 + 1)/sqrt(o**2+4))
def phase_theory(o):
    return (arctan(o) - arctan(o/2))*180/pi
plot(o, gain           , linewidth=7, alpha=0.5, label='gain numeric') # o is omega in rad/s, gain in decibels
plot(o, phase          , linewidth=7, alpha=0.5, label='phase numeric') # phase in degrees
plot(o, gain_theory (o), 'k-'                  , label='gain theory'  )
plot(o, phase_theory(o), 'b-'                  , label='phase theory' )
plot([0], [gain_theory (0)], 'ro', [2], [gain_theory (2)], 'go')
plot([0], [phase_theory(0)], 'ro', [2], [phase_theory(2)], 'go')
grid()
legend()
print(10**(gain_theory(0)/20) , phase_theory (0),'deg', 10**(gain_theory(2)/20), phase_theory(2),'deg')
```

```python
# Let's create our own general Bode plot calculator
# and compare it to the one in the 'signal' package
def my_bode(transfer_function, omega = linspace(0.01, 10,100)):
    return omega,\
           20*log10(abs(transfer_function(1j*omega))),\
           180/pi*angle(transfer_function(1j*omega))
```

```python
H = lambda s: (s+1)/(s+2)
o1, gain1, phase1 = my_bode(H)
print(o1, gain1, phase1)
```

```python
plot(o, gain           , linewidth=7, alpha=0.5, label='gain (signal.bode)') # o is omega in rad/s, gain in decibels
plot(o, phase          , linewidth=7, alpha=0.5, label='phase (signal.bode)') # phase in degrees
plot(o1, gain1, 'k-'                  , label='gain (my_bode)'  )
plot(o1, phase1, 'b-'                  , label='phase (my_bode)' )
legend()
```

## The $z$ transform

## Power series recap:
$$
S_N = \sum_{n=0}^{N} r^n
$$

Multiply by $r$:
$$
rS_N = \sum_{n=0}^{N} r^{n+1}
= \sum_{n=1}^{N+1} r^n=S_{N} + r^{N+1} - 1
$$

$$
(1-r)S_N = 1-r^{N+1}
$$

$$
\boxed{
S_N=\sum_{n=0}^{N}r^n=\frac{1-r^{N+1}}{1-r}
}
$$



If $|r|<1$, then
$$
r^{N+1}\to 0,
$$

so
$$
\boxed{
\sum_{n=0}^{\infty}r^n=\frac{1}{1-r},
\qquad |r|<1
}
$$

## Exercise
Show that if sampled at intervals $T$ the $z$-transform of $x(t) = t$ is
$\frac{Tz}{(z-1)^2}$; 

### Solution


Sampling $x(t)=t$ at intervals $T$ gives
$$
x[n]=x(nT)=nT,\qquad n\ge 0.
$$

Therefore its $z$-transform is
$$
X(z)=\sum_{n=0}^{\infty} nT z^{-n}
=T\sum_{n=0}^{\infty} n z^{-n}.
$$

Start from the geometric series:
$$
\sum_{n=0}^{\infty} r^n=\frac{1}{1-r},\qquad |r|<1.
$$

Differentiate with respect to $r$:
$$
\sum_{n=1}^{\infty} n r^{n-1}
=
\frac{1}{(1-r)^2}.
$$

Multiply by $r$:
$$
\sum_{n=1}^{\infty} n r^n
=
\frac{r}{(1-r)^2}.
$$

Now set
$$
r=z^{-1}.
$$

Then
$$
\sum_{n=0}^{\infty} n z^{-n}
=
\frac{z^{-1}}{(1-z^{-1})^2}.
$$

Hence
$$
X(z)
=
T\frac{z^{-1}}{(1-z^{-1})^2}.
$$

Finally,
$$
T\frac{z^{-1}}{(1-z^{-1})^2}
=
T\frac{z^{-1}}{\left(\frac{z-1}{z}\right)^2}
=
T\frac{z^{-1}z^2}{(z-1)^2}
=
\boxed{
\frac{Tz}{(z-1)^2}
}
$$

with ROC:
$$
\boxed{|z|>1.}
$$

## Exercise
Show that if sampled at intervals $T$ the $z$-transform of $x(t) =e^{-at}$ is $\frac{z}{z-e^{-aT}}$

### Solution

Sampling $x(t)=e^{-at}$ at intervals $T$ gives
$$
x[n]=x(nT)=e^{-anT}=\left(e^{-aT}\right)^n,\qquad n\ge 0.
$$

Therefore its $z$-transform is
$$
X(z)=\sum_{n=0}^{\infty} x[n]z^{-n}
=
\sum_{n=0}^{\infty} \left(e^{-aT}\right)^n z^{-n}.
$$

Rewrite the terms:
$$
X(z)=\sum_{n=0}^{\infty} \left(e^{-aT}z^{-1}\right)^n.
$$

Using the geometric series formula,
$$
\sum_{n=0}^{\infty} r^n=\frac{1}{1-r},\qquad |r|<1,
$$

with
$$
r=e^{-aT}z^{-1},
$$

we get
$$
X(z)=\frac{1}{1-e^{-aT}z^{-1}}.
$$

Equivalently,
$$
\boxed{
X(z)=\frac{z}{z-e^{-aT}}
}
$$

with ROC:
$$
\boxed{
|z|>e^{-aT}
}
$$

## Exercise
Show that if sampled at intervals $T$, the $z$-transform of
$x(t)=\sin \omega t$ is
$$
X_{\sin}(z)=\frac{z\sin\omega T}{z^2-2z\cos\omega T+1},
$$
and the $z$-transform of $x(t)=\cos \omega t$ is
$$
X_{\cos}(z)=\frac{z(z-\cos\omega T)}{z^2-2z\cos\omega T+1}.
$$

### Solution

Using the previous result for $x(t)=e^{-at}$, we may write
$$
\mathcal{Z}\left\{e^{-anT}\right\}
=
\frac{z}{z-e^{-aT}}.
$$

Now take $a=-i\omega$. Then
$$
e^{-aT}=e^{i\omega T},
$$
so
$$
\mathcal{Z}\left\{e^{i\omega nT}\right\}
=
\frac{z}{z-e^{i\omega T}}.
$$

$$
\frac{z}{z-e^{i\omega T}}=\frac{z(z-e^{-i\omega T})}{(z-e^{i\omega T})(z-e^{-i\omega T})}.
$$

Since
$$
(z-e^{i\omega T})(z-e^{-i\omega T})=z^2-2z\cos\omega T+1,$$
we get
$$\frac{z}{z-e^{i\omega T}}=\frac{z(z-e^{-i\omega T})}{z^2-2z\cos\omega T+1}.$$

Now use
$$
e^{-i\omega T}=\cos\omega T-i\sin\omega T.
$$
Therefore
$$
\frac{z}{z-e^{i\omega T}}=\frac{z(z-\cos\omega T+i\sin\omega T)}{z^2-2z\cos\omega T+1}.
$$
But
$$
e^{i\omega nT}=\cos(\omega nT)+i\sin(\omega nT),
$$
so the real part gives the cosine transform:
$$
\boxed{X_{\cos}(z)=\frac{z(z-\cos\omega T)}{z^2-2z\cos\omega T+1}}
$$
and the imaginary part gives the sine transform:
$$
\boxed{X_{\sin}(z)=\frac{z\sin\omega T}{z^2-2z\cos\omega T+1}}
$$

The ROC is
$$
\boxed{|z|>1}
$$
for the causal sampled sequences.


### Alternative Solution
$$
\mathcal{Z}\left\{e^{i\omega nT}\right\}
=
\frac{z}{z-e^{i\omega T}}.
\qquad
\mathcal{Z}\left\{e^{-i\omega nT}\right\}
=
\frac{z}{z-e^{-i\omega T}}.
$$

Using Euler's formulas,
$$
\cos(\omega nT)=\frac{e^{i\omega nT}+e^{-i\omega nT}}{2},
\qquad
\sin(\omega nT)=\frac{e^{i\omega nT}-e^{-i\omega nT}}{2i}.
$$

Therefore
$$
X_{\cos}(z)
=
\frac{1}{2}
\left(
\frac{z}{z-e^{i\omega T}}
+
\frac{z}{z-e^{-i\omega T}}
\right).
$$

Putting the terms over a common denominator,
$$
X_{\cos}(z)
=
\frac{z}{2}
\frac{
(z-e^{-i\omega T})+(z-e^{i\omega T})
}{
(z-e^{i\omega T})(z-e^{-i\omega T})
}.
$$

The numerator becomes
$$
(z-e^{-i\omega T})+(z-e^{i\omega T})
=
2z-\left(e^{i\omega T}+e^{-i\omega T}\right)
=
2z-2\cos\omega T.
$$

The denominator becomes
$$
(z-e^{i\omega T})(z-e^{-i\omega T})
=
z^2-z\left(e^{i\omega T}+e^{-i\omega T}\right)+1
=
z^2-2z\cos\omega T+1.
$$

Hence
$$
\boxed{
X_{\cos}(z)
=
\frac{z(z-\cos\omega T)}{z^2-2z\cos\omega T+1}
}
$$

Similarly,
$$
X_{\sin}(z)=
\frac{1}{2i}
\left(
\frac{z}{z-e^{i\omega T}}-\frac{z}{z-e^{-i\omega T}}
\right).
$$

Putting the terms over a common denominator,
$$
X_{\sin}(z)
=
\frac{z}{2i}
\frac{
(z-e^{-i\omega T})-(z-e^{i\omega T})
}{
(z-e^{i\omega T})(z-e^{-i\omega T})
}.
$$

The numerator becomes
$$
(z-e^{-i\omega T})-(z-e^{i\omega T})
=
e^{i\omega T}-e^{-i\omega T}
=
2i\sin\omega T.
$$

Therefore
$$
\boxed{
X_{\sin}(z)
=
\frac{z\sin\omega T}{z^2-2z\cos\omega T+1}
}
$$
