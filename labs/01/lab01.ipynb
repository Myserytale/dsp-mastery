```python
import matplotlib.pyplot as plt
x = [0, 1]
y = [1, 2]
x, y = [0, 1], [1, 2]
plt.plot(x, y)
plt.show() # to be used in scripts (not in Jupyterlab)
plt.ion() # to be used in terminal for interactive mode (not in Jupyterlab)
```

```python
N = 10
impulse = [0]*N
impulse[0] = 1
impulse
```

```python
np.roll(impulse, N//2)
```

```python
plt.stem(impulse)
```

```python
n = np.arange(N) - N//2
n
```

```python
impulse = np.zeros(N)
impulse[0] = 1
impulse
```

```python
plt.stem(n,impulse[n])
```

```python
step = np.zeros(N)
step[:N//2] = 1
step
plt.stem(n, step[n])
```
